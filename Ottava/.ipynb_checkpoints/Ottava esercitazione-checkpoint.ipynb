# Ottava esercitazione

In questa esercitazione andiamo a ottimizzare con un metodo variazionale lo stato fondamentale di una particella quantistica in una dimensione soggetta a una doppia buca di potenziale del tipo:  $$ V(x) = x^4 - \frac{5}{2}x^2 $$

L'algoritmo che ho sviluppato utilizza una funzione d'onda di prova data dalla somma di due gaussiane parametrizzate dai parametri $\mu$ e $\sigma$ in modo tale che approssimi la soluzione esatta (non calcolabile analiticamente) : $$
\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$$

Il codice contiene al suo interno una funzione in grado di campionare il modulo quadro della funzione d'onda di prova che sfrutta l'algoritmo di Metropolis con una probabilità di transizione uniforme. Grazie a questa funzione è possibile stimare il valor medio dell'hamiltoniana. 


In una seconda parte del codice invece è stata implementata una funzione che sfrutta il "simulated annealing", ovvero la ricerca dello stato a energia minima tramite un lento raffreddamento (fittizio) del sistema. Siccome il sistema in questione non dipende dalla temperatura, interpretiamo il valor medio di H come l'energia di un sistema statistico a una certa temperatura T fittizia e muoviamo i parametri variazionali con accettazione data dal peso di Boltzmann: $$ P=exp(-\beta (L(x')-L(x)))$$

Tale probabilità è facilmente campionabile con l'algoritmo di Metropolis!

Osserviamo che con questo procedimento di fatto si realizza una catena di Markov dipendente dal tempo.



**Soluzione numerica con equazione matriciale**

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

print("Energia stato fondamentale: ", E[0])
print("Energia primo stato eccitato: ", E[1])
print("Energia secondo stato eccitato: ", E[2])
```

### Ricerca parametri variazionali

Partiamo con un valore di $\beta = \frac{1}{K_b T}$ pari a 0.1 ed effettuiamo 200 passi di simulated annealing aumentando $\beta$, e quindi abbassando la temperatura, di 0.2 per ogni passo. 

Conviene lavorare in termini di $\beta$ e non di $T$ perchè aumentando il suo valore andiamo ad esplorare in maniera molto più densa gli stati a bassa temperatura. A bassa temperatura infatti vengono accettate solo mosse che ci avvicinano al ground state, cosa che invece non succede ad alte temperature dove tutte le mosse sono equiprobabili.

```python
fig, (en, ms) = plt.subplots(1,2)
fig.set_figheight(6)
fig.set_figwidth(16)

N=201                         
temp = np.arange(N) 


beta , progaveE, errE, mu, sigma = np.loadtxt("Variational.out", usecols=(0,1,2,3,4), delimiter=' ', unpack='true')
en.errorbar(temp,progaveE,yerr=errE, color='black', label="<H>")
ms.plot(mu,sigma,color='black')

en.plot([0,201],[E[0],E[0]],color='red', label="Energia ground state")
ms.plot(mu[200], sigma[200], 'o', color='red', label='Valori finali')

en.set_xlabel('SA steps')
en.set_ylabel('Energia')
en.set_title('Ricerca parametri variazionali')
en.grid(True)
en.legend();

ms.set_xlabel('$\mu$')
ms.set_ylabel('$\sigma$')
ms.set_title('Traiettoria parametri variazionali')
ms.grid(True)
ms.legend();

plt.show();

print("Mu: ", mu[200])
print("Sigma: ", sigma[200])
```

### Simulazione con parametri fissati

Con questi valori dei parametri variazionali si osserva che, dopo un certo numero di passi montecarlo, il valor medio di H converge a un valore prossimo a quello del ground state ma comunque fuori di 3 deviazioni standard. Ciò è dovuto alla finitezza temporale delle simulazioni che sfruttano il simulated annealing: esse tendenzialmente ci portano in uno stato molto vicino al ground state ma non nel reale ground state.

```python
funz = np.exp(-((x-mu[200])**2)/(2*(sigma[200]**2))) + np.exp(-((x+mu[200])**2)/(2*sigma[200]**2))


fig, (H, histo) = plt.subplots(1,2)
fig.set_figheight(6)
fig.set_figwidth(16)


block , progaveE, errE = np.loadtxt("ene.out", usecols=(0,1,2), delimiter=' ', unpack='true')
H.errorbar(block,progaveE,yerr=errE, color='black')

H.set_xlabel('#blocks')
H.set_ylabel('<H>')
H.set_title('Valor medio hamiltoniana con parametri variazionali fissati')
H.grid(True)
H.plot([0,100],[E[0],E[0]],color='red')

X , p= np.loadtxt("density.out", usecols=(0,1), delimiter=' ', unpack='true')
histo.scatter(X, p/np.sqrt(2), color='black', marker='.', label="Simulazione")

histo.plot(x, ((psi[0])**2)*2, color='blue', label="Soluzione numerica eq matriciale")
histo.plot(x, (funz**2)/np.sqrt(2), color='red', label="Curva analitica")

histo.set_xlabel('x')
histo.set_ylabel('Prob density')
histo.set_title('Densità di probabilità')
histo.grid(True)
histo.legend();

histo.set_xlim((-3,3))
histo.set_ylim((-0.5,2))



plt.show();
```
