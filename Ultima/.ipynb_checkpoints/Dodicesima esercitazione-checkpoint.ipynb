# Dodicesima esercitazione

In questa esercitazione andiamo a costruire una rete neurale in grado di riconoscere e distinguere tra 10 cifre scritte a mano comprese tra 0 e 9. Affrontiamo questo problema inizialmente con una rete neurale composta da layers di tipo Dense() per semplicità, successivamente andiamo a cambiare l'architettura utilizzando dei layer convoluzionali, che di fatto rappresentano la scelta migliore per questo problema.

Per gli esercizi 12.1 e 12.2 utilizziamo delle cifre scritte a mano che appartengono al dataset di MNIST. Tali cifre si presentano come immagini 28x28 in cui ogni pixel può assumere 256 gradazioni diverse di grigio.
Nell'esercizio 12.3 invece utilizziamo delle cifre scritte a mano da me utilizzando l'applicativo gimp.

### Esercizio 12.1

```python
import tensorflow as tf
from tensorflow import keras
import os
import numpy as np
import matplotlib.pyplot as plt
seed=0
np.random.seed(seed) 
tf.random.set_seed(seed)
```

```python
from keras.datasets import mnist

# dimensioni immagine
img_rows, img_cols = 28, 28 # numero di pixel
# output
num_classes = 10 # 10 cifre

# dati scaricati da mnist divisi tra test e train 
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
```

```python
# reshape data
X_train = X_train.reshape(X_train.shape[0], img_rows*img_cols)
X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)
print('X_train shape:', X_train.shape)
print('X_test shape:', X_test.shape)
print()

# cast floats to single precision
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# rescale data in interval [0,1]
X_train /= 255
X_test /= 255


print('esempio di un dato con label', Y_train[20])
# matshow: display a matrix in a new figure window
plt.matshow(X_train[20,:].reshape(28,28),cmap='binary')
plt.show()

# convert class vectors to binary class matrices, e.g. for use with categorical_crossentropy
Y_train = keras.utils.to_categorical(Y_train, num_classes)
Y_test = keras.utils.to_categorical(Y_test, num_classes)
print('... e con label', Y_train[20], 'dopo utilizzo di to_categorical')
print()
print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
```

NB: to_categorical modifica la forma di un array trasformandolo in una matrice con un numero di righe pari alla dimensione del vettore e un numero di colonne pari al numero di classi.

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout

def create_DNN():
    # instantiate model
    model = Sequential()
    # add a dense all-to-all relu layer
    model.add(Dense(400,input_shape=(img_rows*img_cols,), activation='relu'))
    # add a dense all-to-all relu layer
    model.add(Dense(100, activation='relu'))
    # apply dropout with rate 0.5
    model.add(Dropout(0.5))           #Serve a evitare overfitting spegnendo alcuni neuroni
    # soft-max layer
    model.add(Dense(num_classes, activation='softmax'))
    
    return model

```

A questo punto creo tre diversi modelli che utilizzano gli ottimizzatori SGD, Adagrad e Adadelta per fare un confronto sulle performance. Ho sperimentato un po' con il numero di epoche e alla fine ho scelto di utilizzarne 20.

```python
from keras.optimizers import SGD, Adam, RMSprop, Adagrad, Adadelta, Adam, Adamax, Nadam

def compile_model1():           
    # crea modello
    model1=create_DNN()
    # compila il modello
    model1.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=SGD(),
                  metrics=['acc'])
    return model1

def compile_model2():
    
    model2=create_DNN()
    model2.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=Adagrad(),
                  metrics=['acc'])
    return model2

def compile_model3():
    
    model3=create_DNN()
    model3.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=Adadelta(),
                  metrics=['acc'])
    return model3

print('Modelli compilati e pronti per il training.')
```

```python
# parametri training
batch_size = 32      

# creo due reti neurali che differiscono per l'ottimizzatore
model1_DNN = compile_model1()
model2_DNN = compile_model2()
model3_DNN = compile_model3()

```

```python
# training rete 1
history1 = model1_DNN.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=20,
          verbose=1,
          validation_data=(X_test, Y_test))
```

```python
# training rete 2
history2 = model2_DNN.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=20,
          verbose=1,
          validation_data=(X_test, Y_test))
```

```python
# training rete 3
history3 = model3_DNN.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=20,
          verbose=1,
          validation_data=(X_test, Y_test))
```

```python
score1 = model1_DNN.evaluate(X_test, Y_test, verbose=1)
score2 = model2_DNN.evaluate(X_test, Y_test, verbose=1)
score3 = model3_DNN.evaluate(X_test, Y_test, verbose=1)

print()
print('Test loss SGD:', score1[0])
print('Test accuracy SGD:', score1[1])

print('Test loss Adagrad:', score2[0])
print('Test accuracy Adagrad:', score2[1])

print('Test loss Adadelta:', score3[0])
print('Test accuracy Adadelta:', score3[1])

fig, acc=plt.subplots(1,3, figsize=(15,6))

acc[0].plot(history1.history['acc'])
acc[0].plot(history1.history['val_acc'])
acc[0].set_ylabel('model accuracy')
acc[0].set_xlabel('epoch')
acc[0].legend(['train', 'test'], loc='best')
acc[0].set_title('SGD accuracy')
acc[0].grid(True)

acc[1].plot(history2.history['acc'])
acc[1].plot(history2.history['val_acc'])
acc[1].set_ylabel('model accuracy')
acc[1].set_xlabel('epoch')
acc[1].legend(['train', 'test'], loc='best')
acc[1].set_title('Adagrad accuracy')
acc[1].grid(True)

acc[2].plot(history3.history['acc'])
acc[2].plot(history3.history['val_acc'])
acc[2].set_ylabel('model accuracy')
acc[2].set_xlabel('epoch')
acc[2].legend(['train', 'test'], loc='best')
acc[2].set_title('Adadelta accuracy')
acc[2].grid(True)
plt.show()

fig, loss=plt.subplots(1,3, figsize=(15,6))

loss[0].plot(history1.history['loss'])
loss[0].plot(history1.history['val_loss'])
loss[0].set_ylabel('model loss')
loss[0].set_xlabel('epoch')
loss[0].legend(['train', 'test'], loc='best')
loss[0].set_title('SGD loss')
loss[0].grid(True)

loss[1].plot(history2.history['loss'])
loss[1].plot(history2.history['val_loss'])
loss[1].set_ylabel('model loss')
loss[1].set_xlabel('epoch')
loss[1].legend(['train', 'test'], loc='best')
loss[1].set_title('Adagrad loss')
loss[1].grid(True)

loss[2].plot(history3.history['loss'])
loss[2].plot(history3.history['val_loss'])
loss[2].set_ylabel('model loss')
loss[2].set_xlabel('epoch')
loss[2].legend(['train', 'test'], loc='best')
loss[2].set_title('Adadelta loss')
loss[2].grid(True)
plt.show()
```

Dai grafici si nota che tra i tre ottimizzatori lo stochastic gradient descent consente di ottenere le performance migliori sia in termini di precisione che di loss, mentre adadelta risulta essere quello meno performante. Con adagrad invece si ottengono performance simili a SGD.

**Risultati SGD**

### Esercizio12.2

Adesso modifichiamo completamente l'architettura del nostro algoritmo genetico utilizzando una rete convoluzionale per aumentarne la precisione. Vogliamo sfruttare l'esistenza di correlazioni spaziali (locali) tra i pixel delle immagini e il fatto che vi sia un'invarianza traslazionale.

Per prima cosa diamo una nuova forma ai dati:

```python
from keras.layers import Flatten, Conv2D, MaxPooling2D

# reshape data, depending on Keras backend
if keras.backend.image_data_format() == 'channels_first':
    X_train = X_train.reshape(X_train.shape[0], 1, img_rows, img_cols)
    X_test = X_test.reshape(X_test.shape[0], 1, img_rows, img_cols)
    input_shape = (1, img_rows, img_cols)
else:
    X_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1)
    X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
    input_shape = (img_rows, img_cols, 1)
    
print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
print()
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
```

```python
def create_CNN():
    
    model = Sequential()
    
    model.add(Conv2D(10, kernel_size=(5, 5),
                     activation='relu',
                     input_shape=input_shape))
    
    model.add(MaxPooling2D(pool_size=(2,2)))
    
    model.add(Conv2D(16, kernel_size=(5,5),
                     activation='relu'))
    
    model.add(MaxPooling2D(pool_size=(2,2)))
    
    
    model.add(Flatten())     #serve ad appiattire
    
    model.add(Dropout(0.5))    # spegne alcuni neuroni
    
    model.add(Dense(num_classes, activation='softmax'))
    
      
    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer='SGD',
                  metrics=['acc'])
    return model
```

```python
batch_size = 32
epochs = 20

# create the deep conv net
model_CNN=create_CNN()

# train CNN
history=model_CNN.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, Y_test))

# evaliate model
score = model_CNN.evaluate(X_test, Y_test, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
predictions = model_CNN.predict(X_test)

X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols,1)

plt.figure(figsize=(15, 15)) 
for i in range(10):    
    ax = plt.subplot(2, 10, i + 1)    
    plt.imshow(X_test[i, :, :, 0], cmap='gray')    
    plt.title("Digit: {}\nPredicted:    {}".format(np.argmax(Y_test[i]), np.argmax(predictions1[i])))    
    plt.axis('off') 
plt.show()
```

Il tempo di esecuzione su venti epoche è più lungo rispetto al modello precedente, ma il risultato è ottimale. Questa rete neurale è in grado di predirre con precisione tutte le dieci cifre che le abbiamo passato.
Ho scelto di utilizzare SGD come ottimizzatore basandomi sui risultati precedenti.

Riporto ora i grafici della precisione e della loss in funzione del numero di epoche.

```python
score = model_CNN.evaluate(X_test, Y_test, verbose=1)
print()
print('Test loss CNN:', score[0])
print('Test accuracy CNN:', score[1])

fig, acc=plt.subplots(1,2, figsize=(15,6))

acc[0].plot(history.history['acc'])
acc[0].plot(history.history['val_acc'])
acc[0].set_ylabel('model accuracy')
acc[0].set_xlabel('epoch')
acc[0].legend(['train', 'test'], loc='best')
acc[0].set_title('CNN accuracy')
acc[0].grid(True)

acc[1].plot(history.history['loss'])
acc[1].plot(history.history['val_loss'])
acc[1].set_ylabel('model accuracy')
acc[1].set_xlabel('epoch')
acc[1].legend(['train', 'test'], loc='best')
acc[1].set_title('CNN loss')
acc[1].grid(True)
```

### Esercizio 12.3

Adesso testiamo la CNN appena sviluppata su dieci cifre scritte a mano da me utilizzando l'applicativo gimp.

```python
from PIL import Image
import os


fig = plt.figure(figsize =(20,8))

for number in range(10) :
  
  digit_filename = str(number)+'.png'
  digit_in = Image.open(digit_filename).convert('L') 

  ydim, xdim = digit_in.size
  
  pix=digit_in.load();
  data = np.zeros((xdim, ydim))
  for j in range(ydim):
      for i in range(xdim):
          data[i,j]=pix[j,i]

  data /= 255
  data = data.reshape(1,xdim*ydim)

  if keras.backend.image_data_format() == 'channels_first':
      data = data.reshape(data.shape[0], 1, img_rows, img_cols)
      input_shape = (1, img_rows, img_cols)
  else:
      data = data.reshape(data.shape[0], img_rows, img_cols, 1)
      input_shape = (img_rows, img_cols, 1)

  predict = model_CNN.predict(data)
  ax = plt.subplot(2, 10, number + 1)    
  data = data.reshape(xdim,ydim) 
  plt.imshow(data, cmap='gray')    
  plt.title("Digit: {}\nPredicted: {}".format(number,np.argmax(predict)))
  plt.axis('off') 

plt.show()
```

L'algoritmo è in grado di decifrare anche queste cifre scritte (male) a mano!
