### Esercizio 02.1

Vogliamo calcolare numericamente il valore del seguente integrale di cui conosciamo il valore:
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

**Campionamento uniforme**

Come primo metodo utilizziamo un campionamento uniforme nell'intervallo $[0,1]$.

La stima dell'integrale ci è data da $I=\frac{1}{N}\sum_{i=1}^N f(x_i)$, dove $x_i$ sono variabili distribuite uniformemente in $[0,1]$, $f$ è la funzione integranda e $N$ è il numero di estrazioni.

Anche in questo caso sfruttiamo il data-blocking per ridurre le correlazioni tra le misure con un totale di $M=100000$ lanci divisi in $N=100$ blocchi.

Riportiamo nel grafico la stima dell'integrale con rispettiva incertezza in funzione del numero di blocchi.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
            
N=100                           
x = np.arange(N) 

sum_prog, err_prog = np.loadtxt("1.txt", usecols=(0,1), delimiter=' ', unpack='true')

plt.errorbar(x,sum_prog,yerr=err_prog,fmt='-')
plt.xlabel('#blocks')
plt.ylabel('I')
plt.grid(True)
plt.show()
```

La stima dell'integrale rientra praticamente entro una sigma dal valore vero ed è un risultato più che accettabile...ma adesso cerchiamo di dare una stima più precisa con un metodo più sottile.

**Importance sampling**

Sapendo che un certo integrale $I$ tra 0 e 1 è pari a 
$$
I=\int_0^1 f(x)dx = \int_0^1 g(x)p(x)dx
$$
dove $p(x)$ è una distribuzione di probabilità

se scegliamo $p(x)=1$ otteniamo la stessa stima ottenuta con il campionamento uniforme, ma se invece scegliamo $p(x)$ in modo che sia simile all'integranda (ad esempio prendendo il suo sviluppo di Taylor) si ottiene una varianza più piccola di uno o più ordini di grandezza e una stima molto più precisa dell'integrale.

Nel nostro caso ho scelto la seguente distribuzione di probabilità
$$
p(x)=2(1-x)
$$
poichè approssima in maniera abbastanza precisa l'andamento della $f(x)$ nell'intervallo di integrazione. Per campionare punti distribuiti secondo questa distribuzione ho utilizzato il metodo dell'inversione della cumulativa.

Quindi una volta campionati questi punti la stima dell'integrale avviene con lo stesso metodo di prima, ovvero
$$
I=\frac{1}{N}\sum_{i=1}^N g(x_i)
$$
con $g(x)=\frac{\pi}{4}\frac{cos(\pi x/2)}{1-x}$

```python
sum_prog, err_prog = np.loadtxt("2.txt", usecols=(0,1), delimiter=' ', unpack='true')

plt.errorbar(x,sum_prog,yerr=err_prog)
plt.xlabel('#blocks')
plt.ylabel('I')
plt.grid(True)
plt.show()
```

Si può vedere come la stima dell'integrale sia molto più fine e perfettamente compatibile con il valore vero.

### Esercizio 02.2

1) In questa prima parte dell'esercizio simuliamo $10^4$ volte un random walk tridimensionale di $100$ passi discreti di lunghezza $a=1$ in direzione casuale $x, y,$ o $z$ su un reticolo cubico partendo sempre dall'origine.

Ci interessa verificare l'andamento della quantità $D=\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ in funzione del numero di passi effettuati $i$ e per farlo andiamo a ripetere la simulazione 100 volte per ciascuno dei 100 blocchi di cui poi prendiamo il valor medio di $D$ per ciascun passo.

L'andamento previsto per un processo di diffusione come un random walk è $f(i)=k\sqrt{i}$

```python
steps, distance, err = np.loadtxt("discreto.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
rad=np.sqrt(steps)

plt.errorbar(steps,distance,yerr=err,color='violet')
plt.plot(steps,rad,'o',label=r'$\sqrt{i}$')

plt.xlabel('Steps')
plt.ylabel(r'$\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$')
plt.title('3-dimensional random walk on a cubic lattice')
plt.grid(True)
plt.legend()
plt.show()
```
