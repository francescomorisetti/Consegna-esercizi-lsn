# Nona esercitazione

**Introduzione al problema**

In questa esercitazione abbiamo sviluppato un algoritmo genetico per ottimizzare il problema del "travelling salesman", ovvero minimizzare la distanza da percorrere per visitare un certo numero di città, ciascuna una volta sola, e ritornare nella città di partenza.

L'idea alla base è di creare una popolazione di $n$ individui, ciascuno dei quali contiene una sequenza di interi che rappresenta l'ordine delle città da visitare, e di creare delle generazioni successive utilizzando crossover e mutazioni affinchè nell'ultima generazione vi sia almeno un individuo la cui sequenza minimizzi il percorso.

**Dettagli**

* Numero città: 34

* Numero di cromosomi nella popolazione: 1000

* Probabilità crossover: 60%

* Probabilità mutazioni: 10-20%

* Numero generazioni: 500

Il codice che ho sviluppato è del tipo "solutions generational GA", ovvero alla fine di ogni ciclo viene generato un numero di figli uguale a quello dei genitori e la vecchia popolazione viene completamente rimpiazzata da quella nuova (sotto opportune condizioni). 

La distanza tra le città percorse è stata calcolata con una norma $L^{(1)}$: $$
L^{(1)}(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N) = \sum_{i=1}^{N} |\vec{x}_i-\vec{x}_{i+1}|
$$

Questa quantità rappresenta il fitness di un certo cromosoma ed è il criterio in base a cui viene ordinata la popolazione.

**Parte uno**

In questa simulazione le città sono state posizionate in maniera casuale su una circonferenza di raggio unitario per verificare che l'algoritmo converga in maniera efficace, infatti il percorso minimo per tornare nella città di partenza approssima la circonferenza stessa. 

Nota: è presente una degenerazione dovuta al fatto che la circonferenza può essere percorsa in senso orario o in senso antiorario.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
fig = plt.figure(figsize=(8, 8))
x, y = np.loadtxt("circle/circle.out", usecols=(0,1), delimiter=' ', unpack='true')

plt.plot(x, y,color='black')
plt.scatter(x, y, marker='o', color='red')
plt.plot(x[0],y[0],marker='o', color='blue', label='Prima (e ultima) città')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Miglior percorso')
plt.legend()
plt.grid(True)

plt.show()
```

Nel grafico seguente si osserva come la distanza da percorrere venga minimizzata già intorno alla cinquantesima generazione e rimanga tale al crescere del numero di generazioni.

La media è stata fatta sulla prima metà di popolazione (ovvero quella migliore) per ciascuna generazione.

```python
fig = plt.figure(figsize=(8, 8))
n, L = np.loadtxt("circle/L1.out", usecols=(0,1), delimiter=' ', unpack='true')

plt.plot(n, L,color='black')
plt.xlabel('Generation')
plt.ylabel('$Mean L1$')
plt.title('Media fitness sui cromosomi migliori')
plt.grid(True)

plt.show()
```
