# Undicesima esercitazione

In questa esercitazione andiamo a costruire delle reti neurali con Keras con un approccio di tipo supervised learning per affrontare il problema della regressione. 

Il supervised learning ci consente di fittare funzioni a partire da un dataset $D=(X,Y)$, un modello $f(X,p)$ e una funzione costo che ci consente di capire se il modello funziona oppure no.

Il dataset $D$ viene diviso in due gruppi, $D_(train)$ e $D_(test)$, in modo da "allenare" il modello sui dati di training per minimizzare la funzione costo e successivamente valutare la performance sui dati di test. Generalmente i dati di training rappresentano circa il 90% dei dati totali, io ho scelto di aumentare questo valore.

L'idea alla base delle reti neurali arriva proprio dalla biologia, esse infatti sono costituite da una fitta rete di neuroni artificiali che ricevono e inviano dei segnali in base a come vengono stimolati. Un neurone artificiale riceve uno o più inputs $(x_1,x_2,...)$ che vengono pesati con dei pesi $(w_1,w_2,...)$ e fornisce un output a cui si somma un bias $w_0$ che dipende dalla funzione di attivazione specifica di quel neurone.
La funzione di attivazione può avere andamenti molto diversi e la sua regolarità determina in maniera cruciale il funzionamento della rete neurale.


### Esercizio 11.1

In questo esercizio vogliamo fittare la funzione lineare $f(x)=2x+1$ con $x \in [-1,1]$

Questo compito risulta perfetto per una rete costituita da un singolo neurone poichè, come visto precedentemente, l'input di un neurone viene pesato e biasato in maniera analoga a come la variabile $x$ viene moltiplicata per il coefficiente angolare e sommata all'intercetta.

Scegliamo di utilizzare i seguenti parametri:
- numero dati training: 1000
- numero dati test: 50
- funzione costo: MSE
- ottimizzatore: Stochastic gradient descent
- metrica: MSE
- numero epoche: 50
- rumore gaussiano sui dati con $\sigma =0$

```python
# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect

import numpy as np
import matplotlib.pyplot as plt

# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 1000)
x_valid = np.random.uniform(-1, 1, 50)
x_valid.sort()
y_target = m * x_valid + b # ideal (target) linear function

sigma = 0.0 # noise standard deviation, for the moment it is absent
y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(m * x_valid + b, sigma)
```

```python
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,)))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
```

```python
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=50,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
# return weights and biases
model.get_weights()
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear fit')
plt.grid(True)
plt.show()
```

Quindi come previsto con un singolo neurone è possibile fittare funzioni lineari con grande precisione.

### Esercizio 11.2

Adesso ci occuppiamo di fittare la seguente funzione polinomiale:
$$
f(x)=4-3x-2x^2+3x^3
$$
con $x \in [-1,1]$.

Sfruttiamo un modello sequenziale in cui sono presenti 3 layer di neuroni nascosti, fondamentali per poter approssimare una funzione più complicata come quest'ultima

Parametri:

- numero dati training: 500
- numero dati test: 50
- funzione costo: MSE
- ottimizzatore: Stochastic gradient descent
- metrica: MSE
- numero epoche: 100
- rumore gaussiano sui dati con $\sigma =0$
- funzioni di attivazione: relu

```python
# parametri della curva
a = 3 
b = -2 
c = -3
d = 4

# generate training inputs

x_train = np.random.uniform(-1, 1, 500)
x_valid = np.random.uniform(-1, 1, 50)
x_valid.sort()
y_target = a*(x_valid**3) + b*(x_valid**2) + c*x_valid +d

sigma = 0.0 # noise standard deviation, for the moment it is absent
y_train = np.random.normal(a*(x_train**3) + b*(x_train**2) + c*x_train +d, sigma) 
y_valid = np.random.normal(a*(x_valid**3) + b*(x_valid**2) + c*x_valid +d, sigma)
```

```python
model1 = tf.keras.Sequential()
model1.add(Dense(200, input_shape=(1,), activation='relu'))
model1.add(Dense(40, activation='relu'))
model1.add(Dense(30, activation='relu'))
model1.add(Dense(20, activation='relu'))
model1.add(Dense(1, activation='relu'))   

model1.compile(optimizer='sgd', loss='mse', metrics=['mse'])
```

```python
history = model1.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=100,
          shuffle=True, 
          validation_data=(x_valid, y_valid))
```

```python
score = model1.evaluate(x_valid, y_valid, batch_size=32, verbose=0)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
score = model1.evaluate(x_valid, y_target, batch_size=32, verbose=0)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
fig, (hist, mod) = plt.subplots(1,2)
fig.set_figheight(8)
fig.set_figwidth(20)

hist.plot(history.history['loss'], color='red')
hist.plot(history.history['val_loss'], color='black')
hist.set_title('Model loss')
hist.set_ylabel('Loss')
hist.set_xlabel('Epoch')
hist.legend(['Train', 'Test'], loc='best')
hist.grid(True)

x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model1.predict(x_predicted)
mod.scatter(x_predicted, y_predicted,color='r', label='validation data')
mod.plot(x_valid, y_target, label='target', color='black')
mod.set_xlabel('x')
mod.set_ylabel('y')
mod.set_title('Fit')
mod.legend()
mod.grid(True)

plt.show()
```

Si vede come la presenza di layer nascosti incrementi notevolmente la potenza della rete neurale e la renda in grado di fittare funzioni più complicate.



Mostriamo ora come questo tipo di reti neurali perda tutta la sua capacità predittiva se usciamo dal range in cui è stato eseguito il training:

```python
fig = plt.figure(figsize=(8, 8))

x = np.random.uniform(-1, 1.5, 50)
x.sort()
y = a*(x**3) + b*(x**2) + c*x +d


x_predicted = np.random.uniform(-1, 1.5, 100)
y_predicted = model1.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r', label='validation data')
plt.plot(x, y, label='target', color='black')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Fit')
plt.legend()
plt.grid(True)

plt.show()
```

### Esercizio 11.3

Infine estendiamo il modello per fittare una funzione di due variabili:
$$f(x,y) = \sin(x^2+y^2)$$
Nel range $x \in [-3/2,3/2]$ and $y \in [-3/2,3/2]$.

Per questo compito utilizziamo sempre un modello sequenziale ma con due input diversi corrispondenti alle due variabili che vengono poi uniti con la funzione Concatenate.

Parametri:

- numero dati training: 5000
- numero dati test: 100
- funzione costo: MSE
- ottimizzatore: Stochastic gradient descent
- metrica: MSE
- numero epoche: 100
- rumore gaussiano sui dati con $\sigma =0$
- funzioni di attivazione: relu

```python
n_train = 5000 
n_valid = 100

x_train = np.random.uniform(-1.5, 1.5, n_train)
y_train = np.random.uniform(-1.5, 1.5, n_train)

x_valid = np.random.uniform(-1.5, 1.5, n_valid)
y_valid = np.random.uniform(-1.5, 1.5, n_valid)

x_valid.sort()
y_valid.sort()
z_target = np.sin(x_valid**2 + y_valid**2) 

sigma = 0.0 
z_train = np.random.normal(np.sin(x_train**2 + y_train**2) , sigma) 
z_valid = np.random.normal(np.sin(x_valid**2 + y_valid**2) , sigma)
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
fig.set_figheight(8)
fig.set_figwidth(20)

ax = plt.axes(projection='3d')
X, Y = np.meshgrid(x_valid, y_valid)
Z = np.sin(X**2 + Y**2) 

ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('f(x,y)');

plt.show()
```

```python
input1 = keras.layers.Input(shape=(1, ))
input2 = keras.layers.Input(shape=(1,))
merged = keras.layers.Concatenate(axis=1)([input1, input2])
dense1 = keras.layers.Dense(200, input_dim=2, activation='relu', use_bias=True)(merged)
dense2 = keras.layers.Dense(200, input_dim=2, activation='relu', use_bias=True)(dense1)
dense3 = keras.layers.Dense(100, input_dim=2, activation='relu', use_bias=True)(dense2)
dense4 = keras.layers.Dense(50, input_dim=2, activation='relu', use_bias=True)(dense3)

outputs = keras.layers.Dense(1, activation='relu', use_bias=True)(dense4)
model2 = keras.Model(inputs=[input1, input2], outputs=outputs)
```

```python
model2.summary()
```

```python
model2.compile(optimizer=tf.keras.optimizers.SGD(momentum = 0.5, learning_rate = 0.07), loss='mse', metrics=['mse'])
```

```python
history = model2.fit([x_train,y_train], z_train, 
          batch_size=32, epochs=100,
          shuffle=True, 
          validation_split=0.2)
```

```python
score = model2.evaluate([x_valid, y_valid], z_valid, batch_size=32, verbose=0)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
fig = plt.figure(figsize=(8, 8))

plt.plot(history.history['loss'], color='red')
plt.plot(history.history['val_loss'], color='black')
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.grid(True)





plt.show()
```

```python
fig = plt.figure(figsize=(15, 15))

x_predicted = np.random.uniform(-1.5, 1.5, 1000)
y_predicted = np.random.uniform(-1.5, 1.5, 1000)
z_predicted = model2.predict([x_predicted, y_predicted])


mod = plt.axes(projection='3d')
mod.scatter3D(x_predicted, y_predicted, z_predicted, cmap='Greens')

mod.set_xlabel('x')
mod.set_ylabel('y')
mod.set_zlabel('z')
mod.set_title('')


X, Y = np.meshgrid(x_valid, y_valid)
Z = np.sin(X**2 + Y**2) 

mod.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
mod.view_init(10, 20)
mod.grid(True)

```

Si osserva come la potenza predittiva di questo modello diminuisca agli estremi del dominio ma resti molto buona nella parte centrale dove la funzione ha un andamento più semplice. Per funzioni più "complesse" come questa la capacità predittiva del modello dipende molto dalla struttura dei layer nascosti e dalle funzioni di attivazione utilizzate.
