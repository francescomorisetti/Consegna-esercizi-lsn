### Esercizio 07.2



**Funzioni di autocorrelazione**

Dopo aver aggiunto queste correzioni ci interessiamo a dare una stima di quanto devono essere lunghi i blocchi della simulazione per minimizzare le correlazioni. 

Per prima cosa effetuiamo una simulazione con $M=5\times 10^5$ blocchi,  ciascuno composto da un solo passo montecarlo, per ottenere dei valori istantanei dell'energia potenziale per particella dopo una prima fase di equilibrazione. La simulazione viene effettuata  nelle seguenti condizioni termodinamiche:

- stato solido: $\rho^\star = 1.1$, $T^\star = 0.8$ (raggio cut-off: $r_c = 2.2$)
- stato liquido: $\rho^\star = 0.8$, $T^\star = 1.1$ (raggio cut-off: $r_c = 2.5$)
- stato gassoso: $\rho^\star = 0.05$, $T^\star = 1.2$ (raggio cut-off: $r_c = 5.0$)

Una volta ottenuti questi dati calcoliamo la funzione di autocorrelazione per ciascuno stato fisico:

$$
\chi(t)=\frac{\langle U(t)U(t'+t)\rangle-\langle U(t)\rangle \langle U(t'+t)\rangle}{\sigma^2_U}
$$

dove t è l'intervallo tra due misure consecutive espresso in passi montecarlo.

Tale funzione assume valore massimo pari a uno quando $t$ è nullo e tende a zero al crescere di $t$, ovvero quando due misure diventano sempre meno correlate.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

step1, Esol , meanEsol, errEsol = np.loadtxt("es2/solid_epot.dat", usecols=(0,1,2,3), delimiter=' ', unpack='true')
step2, Eliq , meanEliq, errEliq = np.loadtxt("es2/liquid_epot.dat", usecols=(0,1,2,3), delimiter=' ', unpack='true')
step3, Egas , meanEgas, errEgas = np.loadtxt("es2/gas_epot.dat", usecols=(0,1,2,3), delimiter=' ', unpack='true')
```

```python
def autocorrelation(x, nsteps):
    sigma=np.std(x)
    t_max=len(x)
    ac=np.zeros(nsteps)
    
    for dt in range(nsteps):
        sum1=0.0
        sum2=0.0
        sum3=0.0
        tf=t_max-dt
        
        for t in range(tf):
            sum1+=x[t]*x[t+dt]
            sum2+=x[t]
            sum3+=x[t+dt]
        
        ac[dt]=(sum1/tf-sum2*sum3/(tf**2))/(sigma**2)
        
    return ac
```

```python
%%time
tmax=500
time=np.linspace(1,tmax,tmax)

ac_solid = autocorrelation(Esol,500)
ac_liquid = autocorrelation(Eliq,500)
ac_gas = autocorrelation(Egas,500)
```

```python
fig = plt.figure(figsize=(12, 10))

plt.plot(time, ac_solid, color='red', label='Stato solido')
plt.plot(time, ac_liquid, color='blue', label='Stato liquido')
plt.plot(time, ac_gas, color='green', label='Stato gassoso')

plt.xlabel('t (Numero passi montecarlo)')
plt.ylabel('$\chi(t)$')
plt.title('Funzioni di autocorrelazione')
plt.legend()
plt.grid(True)

plt.show()
```

Come previsto la funzione di autocorrelazione decade esponenzialmente in un tempo caratteristico che corrisponde proprio al numero di mosse per singolo blocco che stiamo cercando. Osservando il grafico possiamo estrarre i seguenti valori:

- stato solido: lunghezza blocco$\simeq120$
- stato liquido: lunghezza blocco$\simeq300$
- stato gassoso: lunghezza blocco$\simeq50$

**Data blocking**

A questo punto, per fare un ulteriore controllo su quale deve essere la lunghezza dei singoli blocchi, effettuiamo un calcolo delle incertezze statistiche al variare del numero di lanci per blocco in un range $[10,5000]$ per ciascuna delle tre fasi. 

```python
def error(x,blocks):
    M=len(x)
    L=int(M/blocks)
    ave=np.zeros(blocks)
    
    for i in range(blocks):
        for j in range(L):
            
            ave[i]+=x[j+i*L]
            
        ave[i]=ave[i]/L
    
    error=np.sqrt(np.var(ave)/blocks)
    
    return error
```

```python
%%time

M=5*(10**5)
L = [10, 16, 20, 25, 32, 40, 50, 80, 100, 125, 160, 200, 250, 400, 500, 625, 800, 1000, 1250, 2000, 2500, 3125, 4000, 5000]
n=len(L)
     
solid_error=np.zeros(n)
liquid_error=np.zeros(n)
gas_error=np.zeros(n)

for i in range(n):
    
    N=int(M/L[i])
    solid_error[i]=error(Esol,N)
    liquid_error[i]=error(Eliq,N)
    gas_error[i]=error(Egas,N)
    
```

```python
fig, err=plt.subplots(3,1, figsize=(10,20))

err[0].plot(L,solid_error, color='red')
err[1].plot(L,liquid_error, color='blue')
err[2].plot(L,gas_error, color='green')

err[0].set_xlabel('L')
err[0].set_ylabel('Stat uncertanties')
err[0].set_title('Stato solido')
err[0].grid(True)

err[1].set_xlabel('L')
err[1].set_ylabel('Statistical uncertanties')
err[1].set_title('Stato liquido')
err[1].grid(True)

err[2].set_xlabel('L')
err[2].set_ylabel('Statistical uncertanties')
err[2].set_title('Stato gassoso')
err[2].grid(True)

plt.show()

```

Quello che si vede dai grafici è un progressivo aumento dell'incertezza che tende a saturare a un valore circa costante. Questo valore rappresenta il valore vero dell'errore statistico dovuto all'utilizzo di una catena di Markov.

Per fare un controllo preciso possiamo verificare l'andamento degli errori in un range più corto che contenga i valori di $L$ trovati precedentemente con le funzioni di autocorrelazione.

```python
L = np.linspace(1,600,100)
n=len(L)
     
solid_error=np.zeros(n)
liquid_error=np.zeros(n)
gas_error=np.zeros(n)

for i in range(n):
    
    N=int(M/L[i])
    solid_error[i]=error(Esol,N)
    liquid_error[i]=error(Eliq,N)
    gas_error[i]=error(Egas,N)
    
```

```python
fig, err=plt.subplots(3,1, figsize=(10,20))

err[0].plot(L,solid_error, color='red')
err[1].plot(L,liquid_error, color='blue')
err[2].plot(L,gas_error, color='green')

err[0].set_xlabel('L')
err[0].set_ylabel('Stat uncertanties')
err[0].set_title('Stato solido')
err[0].grid(True)

err[1].set_xlabel('L')
err[1].set_ylabel('Statistical uncertanties')
err[1].set_title('Stato liquido')
err[1].grid(True)

err[2].set_xlabel('L')
err[2].set_ylabel('Statistical uncertanties')
err[2].set_title('Stato gassoso')
err[2].grid(True)

plt.show()
```

Come previsto l'incertezza satura per valori di $L$ circa uguali a quelli trovati con le funzioni di autocorrelazione, fatta eccezione per la fase liquida in cui si vede che anche con blocchi di lunghezza 500 l'incertezza statistica continua a crescere (seppur lievemente).

Per questo motivo conviene scegliere dei blocchi di lunghezza almeno pari a $600$ in modo da non sottostimare l'incertezza statistica.

### Esercizio 07.3

A questo punto aggiungiamo al codice il calcolo della funzione di distribuzione radiale $g(r)$ con $r\in [0,L/2]$.

Questa funzione fornisce la probabilità di trovare una coppia di atomi a una distanza $r$ e assume un andamento tipico per ciascuno stato fisico del gas.
Nell'algoritmo il calcolo della $g(r)$ avviene riempiendo un istogramma in cui ogni specifico bin viene incrementato di 2 ogni volta in cui si trovano due particelle a distanza compresa tra $r$ e $r+dr$.
Infine questo istogramma viene normalizzato con la quantità 
$$
\rho N \frac{4\pi}{3}[(r+dr)^3 - r^3]
$$
dove $\rho$ è la densità del gas e $N$ il numero di atomi.

### Esercizio07.4

Effettuiamo una simulazione di un modello Lennard-Jones di argon utilizzando nostro codice Monte Carlo NVT completo e illustriamo i valori medi (con relative incertezze) di energia potenziale per particella $U/N$, pressione $P$ e funzione di distribuzione radiale $g(r)$ in unità di misura del sistema internazionale.

2. stato liquido: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
3. stato gassoso: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

Effettuiamo anche un confronto tra i valori medi della $g(r)$ ottenuti con l'algoritmo Monte Carlo NVT e con l'algoritmo di dinamica molecolare NVE.

```python
# conversione SI
k_b=1.380649e-23
a_m_u=1.66054e-27

sigma_ar=0.34e-9
epsilon_ar=120.*k_b
m_ar=39.948*a_m_u
```

**Stato solido**

```python
step1, binsol , Grsol, errGrsol = np.loadtxt("es4/solid_gave.dat", usecols=(0,1,2,3), delimiter=' ', unpack='true')
step2, binsolNVE , GrsolNVE, errGrsolNVE = np.loadtxt("NVE/solid/solidNVE_gave.dat", usecols=(0,1,2,3), delimiter=' ', unpack='true')
fig = plt.figure(figsize=(12, 10))

binsol=binsol*sigma_ar
binsolNVE=binsolNVE*sigma_ar

plt.errorbar(binsol, Grsol, yerr=errGrsol, color='blue', label='MC')
plt.errorbar(binsolNVE, GrsolNVE, yerr=errGrsolNVE, color='red', label='MD')
plt.xlabel('r[m]')
plt.ylabel('g(r)')
plt.title('Radial distribution function')
plt.legend()
plt.grid(True)

plt.show()



```

```python
fig, err=plt.subplots(1,2, figsize=(15,8))
block, Esol , aveEsol, errEsol = np.loadtxt("es4/solid_epot.dat", usecols=(0,1,2,3), delimiter=' ', unpack='true')
block, Psol , avePsol, errPsol = np.loadtxt("es4/solid_pres.dat", usecols=(0,1,2,3), delimiter=' ', unpack='true')

aveEsol=aveEsol*epsilon_ar
errEsol=errEsol*epsilon_ar

avePsol=avePsol*epsilon_ar/(sigma_ar**3)
errPsol=errPsol*epsilon_ar/(sigma_ar**3)

err[0].errorbar(block,aveEsol, yerr=errEsol, color='black')
err[1].errorbar(block,avePsol, yerr=errPsol, color='black')


err[0].set_xlabel('Blocks')
err[0].set_ylabel('U/N [J]')
err[0].set_title('Energia potenziale per particella')
err[0].grid(True)

err[1].set_xlabel('Blocks')
err[1].set_ylabel('P [Pa]')
err[1].set_title('Pressione')
err[1].grid(True)

plt.show()
```
