## Sesta esercitazione

Questa esercitazione si occupa di simulare il modello di Ising in una dimensione sfruttando due metodi di campionamento diversi: l'algoritmo di metropolis e il campionamento di Gibbs.

Il modello di Ising nasce per spiegare fenomeni magnetici nella materia e la sua geometria è molto semplice, ogni cella di una griglia (1D, 2D o 3D) rappresenta uno spin e per ciascuno di essi sono possibili solo due orientazioni: up, down.

L'hamiltoniana di un modello con $N$ spin (Assumendo $\mu_B=1$ and $k_B=1$) è data da
$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) 
$$

$s_i$ possono assumere soltanto valori $+1,-1$ e rispettano le condizioni periodiche al contorno $s_{N+1} = s_{1}$.

La funzione di partizione per un modello 1D può essere calcolata esplicitamente e questo risulta molto comodo per dedurre tutti i parametri termodinamici di interesse, difatti vogliamo che il nostro codice fornisca una stima dell'energia interna $U(N,T)$, della capacità termica $C(N,T)$, della magnetizzazione $M(N,T,h)$ e della suscettività magnetica $\chi(N,T)$.


**Campionamento con algoritmo di Metropolis**

Partiamo con il sistema in una configurazione in cui gli spin sono orientati casualmente senza correlazione, ovvero nel limite di temperatura infinita. Scegliamo poi uno spin in maniera random e proviamo a flipparlo. Accettiamo questa mossa con probabilità data dall'algoritmo di metropolis:
$$
A(\nu|\mu)=min[1,exp(-\beta(E_\nu-E_\mu)])
$$

Questo tipo di campionamento ci consente di cercare una configurazione di equilibrio ripetendo il procedimento per tutto il reticolo.

La misura dei parametri termodinamici infatti avviene quando il sistema è arrivato a una temperatura di equilibrio.

**Campionamento di Gibbs**

Il campionamento di Gibbs è un esempio molto semplice di catena di Markov e si differenzia dall'algoritmo di Metropolis per il fatto che ogni nuova mossa proposta viene accettata.

Nel modello di Ising si parte sempre da una configurazione fuori dall'equilibrio e si visitano tutti gli spin ripetutamente assegnandogli dei nuovi valori di spin indipendentemente dai valori che possedevano prima, in tal senso ogni mossa proposta è accettata.

Nello specifico, data 
$$P=\frac{1}{1+exp(\beta(E_u-E_d))}$$

assegnamo valore $+1$ se $r<P$ o valore $-1$ altrimenti, dove $r$ è un numero casuale distribuito uniformemente tra $0$ e $1$.

**Misura parametri termodinamici**

Effettuiamo la simulazione per un sistema di $N=50$ spin, $J=1$ e campo esterno nullo tranne per la misura di $M$ in cui è presente un campo $h=0.02$.

Nei grafici riportiamo l'andamento dei parametri termodinamici per entrambi i metodi di campionamento in funzione della temperatura del sistema in un range $[0.5;2.0]$.

- N blocchi: 100
- N passi per blocco: 100000

```python
fig, (en, susc, heat, mag) = plt.subplots(4)
fig.set_figheight(22)
fig.set_figwidth(10)

points=100
T = np.linspace(0.2,3.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
en.plot(T, e)



temp, aveE , progaveE, errE = np.loadtxt("Metropolis/output.E0", usecols=(0,1,2,3), delimiter=' ', unpack='true')
temp, aveC , progaveC, errC = np.loadtxt("Metropolis/output.C0", usecols=(0,1,2,3), delimiter=' ', unpack='true')
temp, aveX , progaveX, errX = np.loadtxt("Metropolis/output.X0", usecols=(0,1,2,3), delimiter=' ', unpack='true')
temp, aveM , progaveM, errM = np.loadtxt("Metropolis/output.M", usecols=(0,1,2,3), delimiter=' ', unpack='true')

temp, avegE , progavegE, errgE = np.loadtxt("Gibbs/output.E0", usecols=(0,1,2,3), delimiter=' ', unpack='true')
temp, avegC , progavegC, errgC = np.loadtxt("Gibbs/output.C0", usecols=(0,1,2,3), delimiter=' ', unpack='true')
temp, avegX , progavegX, errgX = np.loadtxt("Gibbs/output.X0", usecols=(0,1,2,3), delimiter=' ', unpack='true')
temp, avegM , progavegM, errgM = np.loadtxt("Gibbs/output.M", usecols=(0,1,2,3), delimiter=' ', unpack='true')

en.errorbar(temp,aveE,yerr=errE,fmt='^', color='black')
en.errorbar(temp,avegE,yerr=errgE,fmt='.', color='magenta')
en.set_xlabel('T')
en.set_ylabel('U/N')
en.set_title('Energia interna modello di Ising 1D')
en.grid(True)

ht=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
heat.plot(T, ht)

heat.errorbar(temp,progaveC,yerr=errC,fmt='^', color='black')
heat.errorbar(temp,progavegC,yerr=errgC,fmt='.', color='magenta')
heat.set_xlabel('T')
heat.set_ylabel('C')
heat.set_title('Capacità termica modello di Ising 1D')
heat.grid(True)

x = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
susc.plot(T, x)

susc.errorbar(temp,aveX,yerr=errX, fmt='^', color='black')
susc.errorbar(temp,avegX,yerr=errgX, fmt='.', color='magenta')
susc.set_xlabel('T')
susc.set_ylabel('$\chi$')
susc.set_title('Suscettività magnetica modello di Ising 1D')
susc.grid(True)

h=0.02 #external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
mag.plot(T, M)

mag.errorbar(temp,aveM,yerr=errM, fmt='^', color='black')
mag.errorbar(temp,avegM,yerr=errgM, fmt='.', color='magenta')
mag.set_xlabel('T')
mag.set_xlabel('T')
mag.set_ylabel('M')
mag.set_title('Magnetizzazione modello di Ising 1D con H=0.02')
mag.grid(True)




plt.show()
```
