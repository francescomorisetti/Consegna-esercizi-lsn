# Quarta esercitazione

Lo scopo di questa esercitazione è la simulazione di un sistema molecolare di 108 atomi di argon nelle sue tre fasi basata su un modello Lennard-Jones che utilizza l'algoritmo di Verlet. 

Per ciascuna delle tre fasi il sistema viene fatto partire da una configurazione cristallina per evitare superposizioni, ma, siccome la temperatura iniziale non coincide con la temperatura di equilibrio, il sistema tende a non conservare tale valore di temperatura e a spostarsi verso la sua configurazione più stabile. 

Per questo motivo viene effettuata una fase di equilibrazione per portare il sistema alla temperatura desiderata e poi successivamente viene fatta partire la vera e propria simulazione in cui tener conto di tutti i parametri termodinamici del sistema e farne le statistiche.

### 1)  Stato solido   
$\rho^\star = 1.1$

$T^\star = 0.8$ 

cut-off radius: $r_c = 2.2$


**Fase di equilibrazione:**   

Temperatura di partenza:   $T^* = 1.543$

Facendo partire il sistema da questa temperatura si vede come esso evolva verso $T^* = 0.8$ raggiungendo l'equilibrio in meno di mille lanci.

La fase di equilibrazione viene effettuata prendendo solo valori istantanei in funzione del numero di lanci.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math


plt.plot([0,5000],[0.8,0.8])


blocks, ave, progave, err = np.loadtxt("solid/solid_temp0.dat", usecols=(0,1,2,3), delimiter=' ', unpack='true')
plt.errorbar(blocks,ave,yerr=err)
plt.xlabel('Throws')
plt.ylabel('T')
plt.title('Temperatura argon allo stato solido-Fase di equilibrazione')
plt.grid(True)

plt.show()
```

**Simulazione**

Nei seguenti grafici viene rappresentato l'andamento di energia totale, cinetica, e potenziale (per particella), pressione e temperatura in funzione del numero di blocchi su cui vengono effettuate le medie progressive.

Si nota chiaramente che una volta raggiunta la temperatura di equilibrio il sistema si mantiene stabile e l'energia totale viene conservata.

- numero blocchi: 20
- numero lanci per blocco: 2000

```python

fig, (ekin,epot,etot,pres,temp) = plt.subplots(5)
fig.set_figheight(35)
fig.set_figwidth(10)


blocks, ave, progave, err = np.loadtxt("solid/solid_ekin.dat", usecols=(0,1,2,3), delimiter=' ', unpack='true')
ekin.errorbar(blocks,progave,yerr=err,color="green")
ekin.set_xlabel('#blocks')
ekin.set_ylabel('K/N')
ekin.set_title('Energia cinetica argon allo stato solido')
ekin.grid(True)


blocks, ave, progave, err = np.loadtxt("solid/solid_epot.dat", usecols=(0,1,2,3), delimiter=' ', unpack='true')
epot.errorbar(blocks,progave,yerr=err,color="blue")
epot.set_xlabel('#blocks')
epot.set_ylabel('U/N')
epot.set_title('Energia potenziale argon allo stato solido')
epot.grid(True)


blocks, ave, progave, err = np.loadtxt("solid/solid_etot.dat", usecols=(0,1,2,3), delimiter=' ', unpack='true')
etot.errorbar(blocks,progave,yerr=err,color="red")
etot.set_xlabel('#blocks')
etot.set_ylabel('E/N')
etot.set_title('Energia totale argon allo stato solido')
etot.grid(True)


blocks, ave, progave, err = np.loadtxt("solid/solid_temp.dat", usecols=(0,1,2,3), delimiter=' ', unpack='true')
temp.errorbar(blocks,progave,yerr=err,color="brown")
temp.set_xlabel('#blocks')
temp.set_ylabel('T')
temp.set_title('Temperatura argon allo stato solido')
temp.grid(True)

blocks, ave, progave, err = np.loadtxt("solid/solid_pres.dat", usecols=(0,1,2,3), delimiter=' ', unpack='true')
pres.errorbar(blocks,progave,yerr=err,color="black")
pres.set_xlabel('#blocks')
pres.set_ylabel('P')
pres.set_title('Pressione argon allo stato solido')
pres.grid(True)





plt.show()
```

### 3) Stato gassoso

$\rho^\star = 0.05$

$T^\star = 1.2$ 

cut-off radius: $r_c = 5.0$

**Fase di equilibrazione**

Temperatura di partenza: $T=0.96$

Lo stato gassoso è risultato più complicato da simulare per via dei suoi tempi di equilibrazione molto più lunghi. 

Per oltre 50000 mosse dell'algoritmo di Verlet il sistema continua a muoversi verso la temperatura di equilibrio, motivo per cui ho scelto di utilizzare $10^5$ passi per la fase di equilibrazione.

```python
blocks, ave, progave, err = np.loadtxt("gas/gas_temp0.dat", usecols=(0,1,2,3), delimiter=' ', unpack='true')
plt.errorbar(blocks,ave,yerr=err,color="red")
plt.xlabel('Throws')
plt.ylabel('T')
plt.title('Temperatura argon allo stato gassoso-Fase di equilibrazione')
plt.grid(True)

plt.show()
```

**Simulazione**

Osservando il grafico della temperatura istantanea (fase di equilibrazione) si nota una certa periodicità nelle fluttuazioni anche per tempi molto lunghi. L'utilizzo della statistica a blocchi ci viene in aiuto per minimizzare l'effetto di queste fluttuazioni sul risultato finale, motivo per cui ho scelto di utilizzare dei blocchi di lunghezza maggiore rispetto alle altre due simulazioni, ovvero 20 blocchi da 4000 lanci ciascuno.

```python
fig, (ekin,epot,etot,pres,temp) = plt.subplots(5)
fig.set_figheight(35)
fig.set_figwidth(10)


blocks, ave, progave, err = np.loadtxt("gas/gas_ekin.dat", usecols=(0,1,2,3), delimiter=' ', unpack='true')
ekin.errorbar(blocks,progave,yerr=err,color="green")
ekin.set_xlabel('#blocks')
ekin.set_ylabel('K/N')
ekin.set_title('Energia cinetica argon allo stato gassoso')
ekin.grid(True)


blocks, ave, progave, err = np.loadtxt("gas/gas_epot.dat", usecols=(0,1,2,3), delimiter=' ', unpack='true')
epot.errorbar(blocks,progave,yerr=err,color="blue")
epot.set_xlabel('#blocks')
epot.set_ylabel('U/N')
epot.set_title('Energia potenziale argon allo stato gassoso')
epot.grid(True)


blocks, ave, progave, err = np.loadtxt("gas/gas_etot.dat", usecols=(0,1,2,3), delimiter=' ', unpack='true')
etot.errorbar(blocks,progave,yerr=err,color="red")
etot.set_xlabel('#blocks')
etot.set_ylabel('E/N')
etot.set_title('Energia totale argon allo stato gassoso')
etot.grid(True)


blocks, ave, progave, err = np.loadtxt("gas/gas_temp.dat", usecols=(0,1,2,3), delimiter=' ', unpack='true')
temp.errorbar(blocks,progave,yerr=err,color="brown")
temp.set_xlabel('#blocks')
temp.set_ylabel('T')
temp.set_title('Temperatura argon allo stato gassoso')
temp.grid(True)

blocks, ave, progave, err = np.loadtxt("gas/gas_pres.dat", usecols=(0,1,2,3), delimiter=' ', unpack='true')
pres.errorbar(blocks,progave,yerr=err,color="black")
pres.set_xlabel('#blocks')
pres.set_ylabel('P')
pres.set_title('Pressione argon allo stato gassoso')
pres.grid(True)





plt.show()
```
