# Decima esercitazione

In questa esercitazione riprendiamo in mano l'algoritmo genetico sviluppato per risolvere il problema del commesso viaggiatore e lo paralleliziamo grazie alla libreria MPI ("Message passing interface").
Utilizzando le macchine del laboratorio è possibile sfruttare tutti i loro 10 core e quindi risolvere parallelamente fino a 10 simulazioni per la ricerca del percorso ottimale.
Per far si che la ricerca sia ottimale facciamo in modo che ogni $N_m$ generazioni vengano scambiati gli individui migliori (in maniera casuale) tra i dieci "continenti" che stanno eseguendo la simulazione parallela.

Scegliamo di ottimizzare il percorso che collega 50 capitali americane di cui disponiamo le coordinate scritte su file.

- Numero individui popolazione: 500
- Numero generazioni: 1000
- Numero migrazioni: ogni 50 generazioni
- Probabilità crossover: 60%
- Probabilità mutazioni: 10-20%

### 1) Simulazione con 10 core isolati

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

**Core1**

```python
fig, (dist, L) = plt.subplots(1,2)
fig.set_figheight(7)
fig.set_figwidth(14)


x, y = np.loadtxt("path/path1.out", usecols=(0,1), delimiter=' ', unpack='true')
n, L1 = np.loadtxt("norma/L1.out", usecols=(0,1), delimiter=' ', unpack='true')

L.plot(n, L1,color='black')
L.set_xlabel('Generation')
L.set_ylabel('$Mean L1$')
L.set_title('Media fitness sui cromosomi migliori')
L.grid(True)

dist.plot(x, y,color='black')
dist.scatter(x, y, marker='o', color='red')
dist.plot(x[0],y[0],marker='o', color='blue', label='Prima (e ultima) città')
dist.set_xlabel('x')
dist.set_ylabel('y')
dist.set_title('Miglior percorso')
dist.legend()
dist.grid(True)

plt.show()
```
